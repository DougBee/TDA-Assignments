```python
# Define room descriptions 
rooms = {
    'Living Room': 'You are in the living room. There are antique furniture and a mysterious atmosphere around you.\nThe stairs are to the right, the kitchen is through the door ahead of you, and the dining room is through the door to the right of you.\nWhich direction would you like to go? ',
    'Kitchen': 'You are in the kitchen. You can smell a cake baking in the big old cooker - but there\'s no ghost in this room.',
    'Dining Room': 'You are in the dining room. The room is lightly dimmed - there\'s no ghost in this room.',
    'Stairs': 'You are at the stairs. Press enter to go upstairs',
    'Hallway': 'You are in the hallway on the first floor. Bedroom 1 is to the right, bedroom 2 is to the left, and the stairs to the attic are ahead of you.\nWhich direction would you like to go? ',
    'Bedroom 1': 'You are in bedroom 1. All the furniture is covered - there\'s no ghost in this room.',
    'Bedroom 2': 'You are in bedroom 2 surrounded by lots of candles and a big bed - but no ghost in this room.',
    'Attic': 'Congratulations! You have found the ghost!'
}

# Define room connections and exits 
exits = {
    'Living Room': {'forward': 'Kitchen', 'left': 'Dining Room', 'right': 'Stairs'},
    'Kitchen': {'back': 'Living Room'},
    'Dining Room': {'back': 'Living Room'},
    'Stairs': {'up': 'Hallway', 'down': 'Living Room'},
    'Hallway': {'down': 'Living Room', 'right': 'Bedroom 1', 'left': 'Bedroom 2', 'forward': 'Attic'},
    'Bedroom 1': {'back': 'Hallway'},
    'Bedroom 2': {'back': 'Hallway'},
    'Attic': {'back': 'Hallway'}
}

# Set the starting room
current_room = 'Living Room'
returning_to_living_room = False
returning_to_hallway = False

# Welcome message and get the player's name
print("Welcome to the Haunted House!\nWhat is your name?")
name = input("What is your name? ")
print("Nice to meet you, " + name + ".")

# Instructions 
print("\nInstructions:")
print("There are six rooms in the haunted house. Your job is to find the ghost.\n")
print("You can go forward, left, right, back.\n")
print("If you type 'back', the game will take you back to the last location you were in. Press enter to continue.\n")
input("Press Enter to continue")

# Ask the player if they are ready
print("Are you ready to find the ghost? Type 'Yes' to start or 'No' to quit")
ready_to_start = input("\nAre you ready to find the ghost? Type 'Yes' to start or 'No' to quit: ").lower()

if ready_to_start == 'yes':
    print("\nLet's find the ghost in the haunted house.")
    
    while True:
        # Check if the player is returning to the Living Room or Hallway
        if current_room == 'Living Room' and returning_to_living_room:
            print("You are back in the living room, which direction would you like to go?")
        elif current_room == 'Hallway' and returning_to_hallway:
            print("You are back in the hallway, which direction would you like to go?")
        else:
            # Display current room description
            print(rooms[current_room])

        # Check if the player has found the ghost in the Attic
        if current_room == 'Attic':
            print("Congratulations! You have found the ghost!\nEND OF GAME")
            break

        # Get user input for direction
        user_input = input("Enter 'forward', 'left', 'right', 'back', 'up', or 'quit' to exit: ").lower()

        # Special case for stairs
        if current_room == 'Stairs' and user_input == '':
            current_room = 'Hallway'
            continue

        # Check user input and update current room
        if user_input == 'quit':
            print("Thanks for playing! Goodbye.")
            break
        elif user_input in exits[current_room]:
            previous_room = current_room  # Save the previous room
            current_room = exits[current_room][user_input]

            # Update returning flags
            if user_input == 'back':
                returning_to_living_room = current_room == 'Living Room'
                returning_to_hallway = current_room == 'Hallway'
            else:
                returning_to_living_room = False
                returning_to_hallway = False

        else:
            print("Invalid command. Try again.")
else:
    print("You are not ready for this quest. Goodbye.")
```
